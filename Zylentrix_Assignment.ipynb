```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the dataframes
df1 = pd.read_csv('D:\\Assignment\\students.csv')
df2 = pd.read_csv('D:\\Assignment\\feedback.csv')
df3 = pd.read_csv('D:\\Assignment\\course_activity.csv')
```

```python
merged_df = pd.merge(df1, df2, on="Student_ID", how='outer')
merged_df = pd.merge(merged_df, df3, on="Student_ID", how='outer')
```

```python
#drop the null values
merged_df.dropna()
```

```python
merged_df = merged_df.drop_duplicates()
```

```python
#Average Completion rate across courses
print("The Average Completion rate percentage across courses: ", round(merged_df['Completion_Percentage'].mean(), 2))
```

```python
# Conversion of Data types
merged_df['Enrolment_Date'] = pd.to_datetime(merged_df['Enrolment_Date'], dayfirst=True)
merged_df['Time_Spent_Minutes'] = pd.to_numeric(merged_df['Time_Spent_Minutes'])
merged_df['Time_Spent_Minutes'] = merged_df['Time_Spent_Minutes'].astype('Int64')
```

```python
# Highest and Lowest Average Engagement time
avg_engagement = merged_df.groupby('Course_ID_y')['Time_Spent_Minutes'].mean()
highest_course = avg_engagement.idxmax()
highest_value = avg_engagement.max()
lowest_course = avg_engagement.idxmin()
lowest_value = avg_engagement.min()
print("Highest Average Engagement:" "CourseID-", highest_course, "Time - ", round(highest_value, 2), "mins")
print("Lowest Average Engagement:", "CourseID-", lowest_course, "Time -", round(lowest_value, 2), "mins")
```

```python
#Average Feedback Rating per Course
avg_rating_per_course = merged_df.groupby('Course_ID_y')['Rating'].mean().round(2)
print(avg_rating_per_course)
```

```python
# Engagement based on age and location
age_engagement = merged_df.groupby('Age')['Time_Spent_Minutes'].mean().sort_index()
age_engagement_df = age_engagement.reset_index()
age_engagement_df.columns = ['Age', 'Avg Engagement in Minutes']
print(age_engagement_df)
location_engagement = merged_df.groupby('Location')['Time_Spent_Minutes'].mean().sort_index()
location_engagement_df = location_engagement.reset_index()
location_engagement_df.columns = ['Location', 'Avg Engagement in Minutes']
print(location_engagement_df)
```

```python
#Correlation between Feedback Rating and Completion Percentage
correlation = merged_df['Completion_Percentage'].corr(merged_df['Rating'])
print("Correlation between feedback rating and completion percentage", round(correlation, 2))

#Scatterplot to show the correlation
sns.scatterplot(data=merged_df, x='Rating', y='Completion_Percentage')
plt.title('Feedback Rating Vs Completion Percentage')
plt.xlabel('Feedback rating')
plt.ylabel('Completion_Percentage')
```

```python
#Top 3 student segment based on engagement & satisfaction 
bins = [17, 21, 25, 29, 34]
labels = ['18-21', '22-25', '26-29', '30-34']
merged_df['Age_Group'] = pd.cut(merged_df['Age'], bins=bins, labels=labels)
segment_stats = merged_df.groupby('Age_Group', observed=True)[['Time_Spent_Minutes', 'Rating']].mean()
segment_stats = segment_stats.sort_values(by=['Time_Spent_Minutes', 'Rating'], ascending=False)
top_segment = segment_stats.head(3)
print("Top 3 Students based on engagement and satisfaction")
print(top_segment)
```
